```python
from pycaret.classification import load_model,predict_model
```

```python
saved_final_lr = load_model('Final LR Model sjjgyc')
```

```python
import pandas as pd
test_data = pd.read_csv('test.csv',encoding='utf-8')
```

```python
new_prediction = predict_model(saved_final_lr, data=test_data)
```

```python
new_prediction.head()
```

```python
final_lda = load_model('Final LDA Model sjjgyc')
```

```python
X_lda = final_lda.transform(test_data.drop(columns=['id']))
X_lda
```

```python
# 获取目标变量
targets = predict_model(final_lda,test_data)
targets = targets['prediction_label']
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 设置颜色
colors = ['navy','green','blue','purple']

# target映射
label = {
    0:'low cost',
    1:'medium cost',
    2:'high cost',
    3:'very high cost'
}

# 创建一个图形和子图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 取target唯一值，为数据点分配不同的颜色
for target in targets.unique():
    # 取出所有相同值的target的索引
    indices = targets[targets == target].index
    ax.scatter(X_lda[indices, 0], 
               X_lda[indices, 1], 
               X_lda[indices, 2], 
               c=colors[target], label=f'{label[target]}', marker='o')

ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')

# 添加图例
ax.legend()

plt.title('3D Visualization of LDA-Transformed Data')
plt.show()
```
