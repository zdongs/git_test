```python
import matplotlib.pyplot as plt
import numpy as np

# 矢量字体与图片
plt.rcParams['font.family']=['Microsoft YaHei']
%config InlineBackend.figure_format='svg'
```

## 折线图

```python
# 数据
y = np.random.random(100)
x = np.arange(100)
```

```python
# 折线图 快速上手的封装方案
plt.plot(x,y)
plt.show()
```

### 中文问题

```python
plt.plot(x,y)
plt.title('随机数值')
plt.show()
```

```python
fig, ax = plt.subplots()  # 创建一个figure内部包含一个Axes
ax.plot(x,y);  # 在axes上绘制数据
ax.grid()  # 网格线
```

```python
np.random.seed(19680801)  # 随机数种子
data = {'a':np.arange(50),
        'c':np.random.randint(0,2000,50),
        'd':np.random.randn(50)}  # randn 符合正态分布的随机数
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig,ax = plt.subplots(figsize=(5,2.7),layout='constrained')
ax.scatter('a','b',c='c',s='d',cmap = 'cool',data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
# plt.show()
```
